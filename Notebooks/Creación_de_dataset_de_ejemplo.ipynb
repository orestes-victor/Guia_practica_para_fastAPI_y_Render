```python
import pandas as pd
import numpy as np
from faker import Faker
```

# Creacion de nuestro dataset de ejemplo

```python
# Configurar la semilla para reproducibilidad
np.random.seed(42)

# Crear una lista de nombres de usuarios
usuarios = [format(i) for i in range(1, 10001)]

# Crear un generador de nombres de usuario ficticios
fake = Faker()

# Crear un DataFrame con datos aleatorios para cada usuario y año
data = {
    'Usuario': [fake.user_name() for _ in range(10000)],
    'user_ID': np.random.randint(1000, 9999, size=10000),
    'Estatura (cm)': np.random.uniform(150, 190, size=10000).round(2),
    'Edad': np.random.randint(18, 65, size=10000),
    'Opinion': np.random.choice(['positiva', 'negativa'], size=10000),
    'Gastos': np.random.uniform(500, 5000, size=10000).round(2),
    'Tipo de Sangre': np.random.choice(['A+', 'B+', 'AB+', 'O+', 'A-', 'B-', 'AB-', 'O-'], size=10000),
    'Año': np.random.choice(range(1984, 1996), size=10000)
}

df = pd.DataFrame(data, index=usuarios)

# Mostrar las primeras filas del DataFrame
df.sample(10)
```

# Guardamos nuestro dataset de ejemplo en formato JSON

```python
# Supongamos que ya tienes un DataFrame llamado 'df' creado previamente

# Especificar la dirección y nombre del archivo JSON donde se guardará el DataFrame
direccion_archivo = "/content/drive/MyDrive/Ayuda_fastAPI_y_RENDER/01_Dataset/df_usuarios.json"  # Cambia esta línea con la dirección y nombre deseados

# Guardar el DataFrame en un archivo JSON en la dirección especificada
df.to_json(direccion_archivo, orient='index')

# Imprimir un mensaje de confirmación
print(f"DataFrame guardado exitosamente en la siguiente dirección:'{direccion_archivo}'")
```
